```python
#  Forex
forex_pairs = ["AUDUSD", "CADUSD", "EURCHF", "EURUSD", "GBPCHF", "NZDJPY", "NZDUSD"]
results = []

#  Loop
for pair in forex_pairs:
  #  Import Data
  data = import_csv_data(f"./csv_files/4H/{pair}_4H.csv")
  data = clean_data(data)

  #  Prepare Data
  data = prepare_data(data)

  #  Parameters
  ratio = 1
  starting_balance = 1000
  volume = 1
  spread = 1*volume

  #  Run Strategy
  rsi_stoch_strategy = Strategy_mcr_candles(data, starting_balance=starting_balance, volume=volume, ratio=ratio, spread=spread)
  result = rsi_stoch_strategy.run()
  results.append(result)
```
